```python
from manim import *

param = "-v WARNING -r 500,180 -s  --disable_caching Example"
```

```python
np.random.seed(20)
steps = np.random.choice(a=[-1, 0, 1], size=(300000, 1000))
stop = steps.cumsum(0)
end_points = stop[-1] / stop[-1].max()
end_pointsX = end_points[0:499]
end_pointsY = end_points[500:-1]
```

```python
%%manim $param
class Example(Scene):
    def construct(self):
        radius= (end_pointsX*end_pointsX + end_pointsY * end_pointsY)**0.5
        dots = VGroup()
        for  x,y,r in zip(end_pointsX, end_pointsY,radius):
            c= interpolate_color(YELLOW, RED, r)
            dots.add(Dot(color=c,point=[3*x,3*y,0]).scale(3))
        self.add(dots)
```
