```python
%%manim -v WARNING -s -ql -t -r 2000,400 --disable_caching --progress_bar None Example

from manim.mobject.geometry import ArrowTriangleFilledTip

class Example(ZoomedScene):
    def construct(self):
        # Banner for flyingframes.readthedocs.io
        self.camera.frame.scale(2) # needs ZoomedScene
        self.camera.background_color = "#ece6e2" 
        logo_green = "#87c2a5"
        logo_blue = "#525893"
        logo_red = "#e07a5f"        
        logo_black = BLACK
        d= Dot(color= logo_green, radius=0.8).shift(UR*0.6)
        l = VGroup(*[Line(color=logo_blue, stroke_width=70/3) for _ in range(0,3)])
        l.arrange(DOWN,buff=0.45).next_to(d, LEFT).shift(0.3*LEFT)
        l[0::2].shift(0.3*RIGHT)
        l.rotate(30*DEGREES, about_point=d.get_center())
        arc= Arc(radius=1.0, color= logo_red, start_angle=PI/4, angle=PI, stroke_width=70)
        arc.shift(1.75*DOWN+1.15*RIGHT)
        
        logo_all= VGroup(arc,l,d)

        logo_black = "#343434"
        M = MathTex(r"\mathbb{F}").scale(7).set_color(logo_black)
        anim= Tex(r"\textbf{lying}",tex_template=TexFontTemplates.gnu_freeserif_freesans).scale(7).set_color(logo_black).scale(0.75748)

        E = MathTex(r"\mathbb{F}").scale(7).set_color(logo_black)
        ditor= Tex(r"\textbf{rames}",tex_template=TexFontTemplates.gnu_freeserif_freesans).scale(7).set_color(logo_black).scale(0.75748)
        x=VGroup(M,anim).arrange(RIGHT, buff=0.1)
        y=VGroup(E,ditor).arrange(RIGHT,buff=0.1)
        banner=VGroup(x,y,logo_all).arrange(RIGHT,buff=0.5)
        b1=M.get_bottom()
        b2=anim[0][0].get_bottom()
        anim.shift(DOWN*(b2-b1))
        ditor.align_to(M,DOWN)
        self.add(RoundedRectangle(fill_color="#ece6e2",height=banner.get_height()+1, width=banner.get_width()+1.5,fill_opacity=1, stroke_width=0, z_index=0))
        
        self.add(x,y,logo_all)
        logo_all.shift(0.2*DOWN)
```
