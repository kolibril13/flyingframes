```python
from manim import *
from manim_mobject_svg import *
config.media_embed = True
```

```python
t = ValueTracker(2)

g = Square().set_stroke(width = 30)
always_redraw(lambda: g.scale_to_fit_width(t.get_value()))
g.set_color(BLUE)
g
```

```python
param = "-v WARNING -r 500,500   --disable_caching Example"

class Example(Scene):
    def construct(self):
        t.set_value(1)
        self.add( g)
        self.play(t.animate.set_value(10), run_time=1)
%manim $param
```
